### Task 4

Change the values in `pclass` to `first`, `second`, and `third`. Make a new column assigning everyone to an age group based on their ages:

* child (0-12 years)
* teen (13-18 years)
* adult (19-40 years)
* mid-aged (41-65 years)
* old aged (66+ years)

Find the number of people in each class and age group. 

```python
# Change the values in 'Pclass'
df['pclass'] = df['pclass'].replace({1: 'first', 2: 'second', 3: 'third'})
df
```

```python
# Define the age groups and their labels
ages = [0, 12, 18, 40, 65, np.inf]
ageLabels = ['child', 'teen', 'adult', 'mid-aged', 'old aged']

# Create a new column for age groups
df['age.group'] = pd.cut(df['age'], bins=ages, labels=ageLabels)
```

```python
# Find the number of people in each class
# and show the total number of people from each class
classCount = df.groupby(['pclass']).size().reset_index(name = 'count')
classCount
```

```python
# Find the number of people in each class and age group
# and show the total number of people from each age group in each class
peopleCount = df.groupby(['pclass','age.group']).size().reset_index(name = 'count')
peopleCount
```

### Task 5

Create a new column called `"FamilySize"` which combines the two columns `sibsp` and `parch` which stores the number of siblings and parents a person had on the ship respectivly. Find the average survival rate based on family size. What family size was the most fatal? What family size led to a higher chance of survival?

```python
# Create a new column 'FamilySize'
df['family.size'] = df['sibsp'] + df['parch']
```

```python
# Calculate the average survival rate based on family size
survivalRate = df.groupby('family.size')['survived'].mean()

# Find the family size with the lowest and highest survival rate
mostFatalFamilySize = survivalRate.idxmin()
highestSurvivalFamilySize = survivalRate.idxmax()
```

```python
survivalRate.reset_index()
```

```python
print("Average Survival Rate based on Family Size:")
print("\nFamily Size with Most Fatalities: ", mostFatalFamilySize)
print("Family Size with Highest Chance of Survival: ", highestSurvivalFamilySize)
```
